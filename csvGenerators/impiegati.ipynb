```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import random
```

```python
persone = pd.read_csv('persone.csv')
reparti = pd.read_csv('reparti.csv')
persone
```

```python
def getVettoreAssunzione():
    dataAssunzione=[]
    for i in range(0, len(persone)):
        dataAssunzione.append('2016-02-14')
    return  dataAssunzione
```

```python
impiegati=pd.DataFrame()
impiegati['dataAssunzione']=getVettoreAssunzione()
impiegati['mansione']='Scaffalista'
impiegati['livello']=5
impiegati['indirizzoReparto']=reparti['supermercato']
impiegati['nomeReparto']=reparti['nome']
impiegati['codF']=persone['codF']
impiegati
```
