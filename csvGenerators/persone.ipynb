```python
from codicefiscale import codicefiscale
import random 
```

```python
def randomSesso():
    sesso = random.randint(0, 1)
    return 'F' if sesso==0 else 'M'
```

```python
codicifiscali=[]
for i in range(0,len(persone)):
        data=persone["Birthday"][i]
        dataobj=datetime.strptime(data, '%Y-%m-%d')
        compleanno=dataobj.strftime('%d/%m/%Y')
        codicifiscali.append(codicefiscale.encode(surname=persone['Surname'][i], name=persone['GivenName'][i], sex=randomSesso(), birthdate= compleanno, birthplace='Milano'))
```

```python
persone['codF']=codicifiscali
persone
```
